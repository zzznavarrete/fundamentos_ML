```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import matplotlib.pyplot as plt
```

```python
fasion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fasion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover','Dress', 'Coat', 'Sandal', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# Viendo cuantos datos tenemos
train_images.shape
```

```python
# Buscamos un elemento en particular para visualizar.
plt.figure()
plt.imshow(train_images[100])
plt.grid(True)
```

```python
# Normalizamos los pixeles
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
%matplotlib inline

plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(train_images[i], cmap= plt.cm.binary)
    #plt.xlabel(class_names[train_labels[i]])
```

```python
class_names[train_labels[1]]
```

```python
# Definimos nuestro modelo
model = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)), keras.layers.Dense(128, activation = tf.nn.relu), keras.layers.Dense(10,activation = tf.nn.softmax)])
```

```python
# Configuro mi optimizador
model.compile(optimizer = tf.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Ajusto el modelo
model.fit(train_images, train_labels, epochs = 5)
```

```python
# Calculando el accuracy de nuestro modelo
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Accuracy ', test_acc)
```

```python
# Creando predicciones
predictions = model.predict(test_images)
```

```python
plt.figure(figsize = (10,10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(test_images[i], cmap = plt.cm.binary)
    predicted_label = np.argmax(predictions[i])
    true_label = test_labels[i]
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'
    # Descomentar esto para ver la prediccin
    #plt.xlabel('{} ({})'.format(class_names[predicted_label], class_names[true_label]), color = color)
```
